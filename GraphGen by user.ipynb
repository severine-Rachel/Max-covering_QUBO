```python
G2 = nx.Graph()
pointsCovered = []
pointsHH = []
pointsVH = []
pointsHL = []
pointsVL = []

def isHorizontal(edge):
  if G.nodes[edge[0]]["coord"][1] == G.nodes[edge[1]]["coord"][1]:
    return True
  else:
    return False

def isVertical(edge):
  if G.nodes[edge[0]]["coord"][0] == G.nodes[edge[1]]["coord"][0]:
    return True
  else:
    return False

def isVTaller(edge1, edge2):
    dE1 = abs(G.nodes[edge1[1]]["coord"][1] - G.nodes[edge1[0]]["coord"][1])
    dE2 = abs(G.nodes[edge2[1]]["coord"][1] - G.nodes[edge2[0]]["coord"][1])
    if dE1 >= dE2:
      return True
    else:
      return False

def isHTaller(edge1, edge2):
  dE1 = abs(G.nodes[edge1[1]]["coord"][0] - G.nodes[edge1[0]]["coord"][0])
  dE2 = abs(G.nodes[edge2[1]]["coord"][0] - G.nodes[edge2[0]]["coord"][0])
  #print("edge1 = ", edge1, "edge2 = ", edge2, "dE1 = ", dE1, "dE2 = ", dE2)
  if dE1 >= dE2:
    return True
  else:
    return False

def SupH(edge1, edge2):
    return G.nodes[edge1[0]]["coord"][1] > G.nodes[edge2[0]]["coord"][1]

def SupV(edge1, edge2):
    return G.nodes[edge1[0]]["coord"][0] > G.nodes[edge2[0]]["coord"][0]

def InfV(edge1, edge2):
    return G.nodes[edge1[0]]["coord"][0] < G.nodes[edge2[0]]["coord"][0]

def InfH(edge1, edge2):
    return G.nodes[edge1[0]]["coord"][1] < G.nodes[edge2[0]]["coord"][1]

#place the street point between two edges representing the street
for edge1 in G.edges:
  if isHorizontal(edge1):
    print("H 1 edge1= ",edge1)
    for edge2 in G.edges:
      if isHorizontal(edge2) and edge1 != edge2:
        print("H 1 edge2= ",edge2)
        if isHTaller(edge1, edge2):
          print("H 2  edge1= ",edge1)
          if SupH(edge1, edge2):
            print("H 3 edge1= ",edge1)
            for node in G1.nodes:
              if node[1] == G.nodes[edge1[0]]["coord"][1]  and node[0] != G.nodes[edge1[0]]["coord"][0] and node[0] != G.nodes[edge1[1]]["coord"][0]:
                point = (node[0],node[1]-p)
                pointsCovered.append(point)
                print("H 6 point covered = ", point)
                if node not in pointsHH:
                  pointsHH.append(node)
                print("H 4 edge1 = ", edge1)
              if node[1] == G.nodes[edge2[0]]["coord"][1] and not (node[0],node[1]) in pointsCovered and node[0] <= G.nodes[edge2[0]]["coord"][0] and node[0] >= G.nodes[edge2[1]]["coord"][0] and node not in pointsHL:
                pointsHL.append(node)
                print("H 5 edge2 = ", edge2)
              if node[1] == G.nodes[edge2[0]]["coord"][1]  and node[0] != G.nodes[edge2[0]]["coord"][0] and node[0] != G.nodes[edge2[1]]["coord"][0]:
                point = (node[0],node[1]+p)
                pointsCovered.append(point)
                print("H 7 point covered = ", point)
          elif InfH(edge1, edge2):
            print("H 3 edge2= ",edge2)
            for node in G1.nodes:
              if node[1] == G.nodes[edge1[0]]["coord"][1]  and node[0] != G.nodes[edge1[0]]["coord"][0] and node[0] != G.nodes[edge1[1]]["coord"][0]:
                point = (node[0],node[1]+p)
                pointsCovered.append(point)
                print("H 8 point covered = ", point)
                if node not in pointsHH:
                  pointsHL.append(node)
                print("H 4 edge1 = ", edge1)
              if node[1] == G.nodes[edge2[0]]["coord"][1] and not (node[0],node[1]) in pointsCovered and node[0] <= G.nodes[edge2[0]]["coord"][0] and node[0] >= G.nodes[edge2[1]]["coord"][0] and node not in pointsHL:
                pointsHH.append(node)
                print("H 5 edge2 = ", edge2)
  if isVertical(edge1):
    #print("V 1 edge1= ",edge1)
    for edge2 in G.edges:
      if isVertical(edge2) and edge1 != edge2:
        #print("V 1 edge2= ",edge2)
        if isVTaller(edge1, edge2):
          #print("V 2  edge1= ",edge1)
          if SupV(edge1, edge2):
            #print("V 3 edge1= ",edge1)
            for node in G1.nodes:
              if node[0] == G.nodes[edge1[0]]["coord"][0] and node[1] != G.nodes[edge1[0]]["coord"][1] and node[1] != G.nodes[edge1[1]]["coord"][1]  and not (node[0],node[1]) in pointsCovered:
                point = (node[0]-p,node[1])
                pointsCovered.append(point)
                print("V 6 point covered = ", point)
                if node not in pointsVH:
                  pointsVH.append(node)
                #print("V 4 edge1 = ", edge1)
              if node[0] == G.nodes[edge2[0]]["coord"][0]  and node[1] != G.nodes[edge1[0]]["coord"][1] and node[1] != G.nodes[edge1[1]]["coord"][1] and not (node[0],node[1]) in pointsCovered and node not in pointsVL:
                pointsVL.append(node)
                #print("V 5 edge2 = ", edge2)
              if node[0] == G.nodes[edge2[0]]["coord"][0]  and node[1] != G.nodes[edge2[0]]["coord"][1] and node[1] != G.nodes[edge2[1]]["coord"][1]:
                point = (node[0]+p,node[1])
                pointsCovered.append(point)
                print("V 7 point covered = ", point)
G1.add_nodes_from(pointsCovered)
print(pointsCovered)

#edge that represents scope between lidar placement and vertices to cover
for node in pointsVL:
  if (node[0]+p,node[1]) in pointsCovered:
        G1.add_edge(node, (node[0]+p,node[1]))
for node in pointsHL:
  if (node[0],node[1]+p) in pointsCovered:
        G1.add_edge(node, (node[0],node[1]+p))
for node in pointsVH:
  if (node[0]-p,node[1]) in pointsCovered:
        G1.add_edge(node, (node[0]-p,node[1]))
for node in pointsHH:
  if (node[0],node[1]-p) in pointsCovered:
        G1.add_edge(node, (node[0],node[1]-p))

print(pointsVL, pointsVH, pointsHL, pointsHH)
pos = {node: node for node in G1.nodes}
nx.draw(G1, pos, with_labels=True)
nx.draw_networkx_nodes(G1, pos, pointsCovered, node_color='red', )
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xticks(range(0, 40))
plt.yticks(range(0, 40))
plt.show()
```
